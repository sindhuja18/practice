```python
def normalizeUnits(val):
    values = {'MB':1.0/1024, 'GB':1.0, 'TB':1024.0}
    var_type = re.findall(r'([0-9]+)', val)
    nsd_val = val
    if var_type:
        for x in values:
            if x in val.upper():
                nsd_val = str(int(var_type[0])*values[x])+' GB'
    return nsd_val
    
```
