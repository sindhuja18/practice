```python
import pandas as pd
import re
```

```python
variants = pd.read_csv('US_Smartphone_variants - New_list_amazonUS.csv',delimiter = ',')
```

```python
def variant(row):
    row = list(row)
    product_name = row[2]
    value = re.findall('([0-9]+([ ]+|[ ])?(MB|GB|Gb|Mb|gb))', product_name)
    print 'First Extraction', value
    values = {}
    for val in value:
        print 'Looping over values current one is ', val
        variant = val[0].replace(val[2], ' '+val[2].upper())
        num = re.findall(r'([0-9]+)', variant)
        print num, variant
        if num:
            values[int(num[0])] = variant
    print 'total variants ',values
    if values:
        value = values[max(values.keys())]
#     if value:
#         value = value[0][0].replace(value[0][2], ' '+value[0][2].upper())
    else:
        value = ''
    row.append(value)
#     row[5]=value
    return row
```

```python
out_values = pd.DataFrame(list(variants.apply(variant,axis=1)), columns=['buysmaart_id', 'product_name', 'source_product_name', 'source_product_url', 'Mobile_Variants', 'memory_variants', 'Color_variants','memory'])
```

```python
out_values.to_csv('variants_new_Amazon.csv',index = False,sep = ',')
```
